<h2>#1 Gas Prices</h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h2>Line Graph</h2>

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.title('Gas Prices over Time(in USD)', fontdict={'fontweight':'bold'})

plt.plot(gas.Year, gas.USA,'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')

#for country in gas:
    #if country != 'Year':
        #plt.plot(gas.Year, gas[country], marker='.', label=country)
    

print(gas.Year[::3])

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price.png')

plt.show()
```

<h2>#2 Fifa Data</h2>

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

<h2>Histogram</h2>

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins)

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA')

plt.show()
```

<h2>Pie Chart</h2>

```python
fifa.head(5)
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
right
```

```python
labels = ['Left', 'Right']

plt.pie([left, right], labels = labels, autopct='%.2f')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight[0]
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
heavy = fifa.loc[fifa.Weight > 200].count()[0]
heavy
```

```python
weights = [light, light_medium, heavy]
labels = ['Under 125', '125-150', 'Over 200']
explode = (.1, .1, .1)

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels=labels, autopct='%.2f', explode=explode)

plt.show()
```
