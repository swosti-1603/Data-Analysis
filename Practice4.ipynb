```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

<h2>Basic Graph</h2>

```python
x = [2,3,4]
y = [4,6,8]

#Keyword Argument Notation
plt.plot(x,y,label='2x', color='red', linewidth=2, marker='.', markersize=10, markeredgecolor='blue', linestyle='--')

#select interval where point is to be plotted
x2 = np.arange(0,4,0.5)
print(x2)

#Plot part of graph as line
plt.plot(x2[:5], x2[:5]**2, 'b', label='X^2')

#Plot remainder of graph as dot
plt.plot(x2[4:], x2[4:]**2, 'b--')

plt.title('Our First Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,2,4,6,8,10])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.savefig('mygraph.png')

plt.show()
```

<h2>Bar Chart</h2>

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('/')
bars[2].set_hatch('/')

plt.show()
```
